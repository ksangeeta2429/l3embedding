```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa
import librosa.display
import numpy as np
from librosa.core.time_frequency import *
```

```python
#References:
#https://librosa.github.io/librosa/generated/librosa.feature.melspectrogram.html
#https://librosa.github.io/librosa/generated/librosa.core.stft.html#librosa.core.stft
#https://librosa.github.io/librosa/generated/librosa.filters.mel.html#librosa.filters.mel
#https://librosa.github.io/librosa/_modules/librosa/filters.html
#https://librosa.github.io/librosa/_modules/librosa/core/time_frequency.html#mel_frequencies
#kapre: https://github.com/keunwoochoi/kapre/blob/master/kapre/time_frequency.py
```

```python
def get_mel_spacing(n_mels, fmax, fmin=0):
    min_mel = hz_to_mel(fmin, htk=True)
    max_mel = hz_to_mel(fmax, htk=True)
    return max_mel//n_mels
```

```python
sr = 48000
n_mels = 256
fmax = sr/2
#Max Mel is 4016.01917987
mel_spacing = get_mel_spacing(n_mels=n_mels, fmax=fmax)
# 'Center freqs' of mel bands - uniformly spaced between limits
mel_f = mel_frequencies(n_mels + 2, fmin=0, fmax=fmax, htk=True)
print('Original L3 Mel spacing: ', mel_spacing)
print('Equal spaced mels in Hz: ', mel_f)
```

```python
audiofile = '/home/sk7898/l3embedding/test_data/1-97392-A-0.wav'
sr = 16000
n_fft = 2048
n_mels = 128
hop_len = 484

audio, sr = librosa.load(audiofile, sr=sr)
melspec = librosa.feature.melspectrogram(audio, sr=sr, n_fft=n_fft, hop_length=hop_len, n_mels=n_mels, htk=True, fmax=sr//2)
melspec = librosa.logamplitude(melspec, ref_power=np.max)

plt.figure(figsize=(10, 4))
librosa.display.specshow(melspec, x_axis='time', y_axis='mel', sr=sr, fmax=sr//2)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
plt.show()
```
